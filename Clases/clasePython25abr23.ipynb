Ejercicio: funcion: recibe coeficientes de ec. cuadratica, devuelve las raices de la ecuacion.



```python
import math
from math import sqrt,exp
import numpy as np
```

```python
def cuadratica (a,b,c):
  b2=sqrt(pow(b,2)-4*a*c)
  if b2<=0:
    print('No tiene soluciones reales')
  else:
    x1=(-b+(sqrt(pow(b,2)-4*a*c)))/2*a
    x2=(-b-(sqrt(pow(b,2)-4*a*c)))/2*a
    return(x1,x2)
```

```python
cuadratica(1,-5,6)
```
